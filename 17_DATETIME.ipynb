```python
 # Date and DateTime are objects, so manipulating them means working with objects, not plain strings or timestamps.
```

```python
import datetime

d = datetime.date(2004, 8, 25)
print(d)
print("Day : ", d.day)
print("Month : ", d.month)
print("Year : ", d.year)
```

```python
# today's date
d1 = datetime.date.today()
print(d1)
print(d1.year)
# formatting date

d1.strftime("%d/%m/%y")
d1.strftime("%A,%d-%B-%Y")
```

```python
from datetime import datetime

date_time = datetime.fromtimestamp(1887639468)
print(date_time)
print(date_time.time())
print(date_time.date())
```

```python
import datetime

d = datetime.date(2020, 12, 12)
print(d)
print(type(d))
date_str = d.isoformat()
print(date_str)
print(type(date_str))
```

### timedelta

```python
from datetime import datetime, timedelta

d = datetime.now()
print("Today", d)
tomorrow = d + timedelta(days=1, hours=27)
print("Tomorrow  3 pm :", tomorrow)
print("difference in time :", tomorrow - d)
```

### timezone

```python
from datetime import datetime, timezone, timedelta

utc_time = datetime.now(timezone.utc)
print("UTC:", utc_time)

#india's time
ist = timezone(timedelta(hours=5, minutes=30))
ist_time = utc_time.astimezone(ist)
print("India ", ist_time)
```

```python
from datetime import datetime
from pytz import timezone

now_utc = datetime.now(timezone("UTC"))
print(now_utc)
timezones = ['Asia/Kolkata', 'Europe/Kiev', 'America/New_York']
for tzone in timezones:
    now_asia=now_utc.astimezone(timezone(tzone))
    print(now_asia)
```

```python
from datetime import datetime
from zoneinfo import ZoneInfo

now_utc = datetime.now(ZoneInfo("UTC"))
now_india = datetime.now(ZoneInfo("Asia/Kolkata"))
now_ny = datetime.now(ZoneInfo("America/New_York"))

print("UTC:", now_utc)
print("India:", now_india)
print("New York:", now_ny)
```
