# Sets

Sets are collections that store unique elements. They are unordered and unindexed, mutable, and denoted by curly braces `{}`.

## Frozensets

Frozensets are similar to sets but immutable. You can't add, remove, or update elements, but you can perform operations like union, intersection, and difference.

empty set
s1=set() # created using set constructor

To create a set with elements(from exisitng), pass any iterable like a list, tuple, range, or string to the `set()` constructor.


```python
s1=set([3, 4, 5])        # From a list
print(s1)
s2=set((3, 4, 5))        # From a tuple
print(s2)
s3=set("abcabc")         # From a string ➝ {'a', 'b', 'c'}
print(s3)
s4=set(range(4))         # From a range ➝ {0, 1, 2, 3}
print(s4)
```

## Set Methods

### List of Methods
- `len()`
- `add()`
- `remove()`
- `discard()`
- `pop()`
- `clear()`
- `copy()`
- `update()`


```python
set1={2,3,4,6,7,55,6,66,7}
print(set1)
```

- `len()` : returns the length of the set

```python
len(set1)
```

- `add(element)`: Adds a single element to the set if it is not already present.

```python
set1.add(4)
print(set1)
```

- `remove(element)`: Removes the specified element from the set. Raises a KeyError if the element is not found.

```python
set1.remove(3)
print(set1)
```

- `discard(element)`: Removes the specified element from the set if it is present. Does not raise an error if the element is not found.

```python
set1.discard(3)
print(set1)
```

- `pop()`: Removes and returns an random value from the set 

```python
set1.pop()
# set1.pop()
# print(set1)
```

- `clear()`: Removes all elements from the set, leaving it empty.

- `copy()`: Returns a shallow copy of the set.

- `update(iterable)`: Adds elements from an iterable (like a list or another set) to the set.

```python
l1={2,3,4,5}
set1.update(l1)
print(set1)

```

| Operation      | Operator |
|----------------|----------|
| Union          | \|       |
| Intersection   | &        |
| Difference     | -        |

```python
# other methods
set1={2,3,4,5}
set2={4,5,6,7}
print(set1.union(set2))
print(set1.intersection(set2))
print(set1.difference(set2))
print(set1.symmetric_difference(set2))

print
```

```python
# removing duplicates from a list

nums = [1, 2, 2, 3, 3, 4]
unique_nums = list(set(nums))  # ➝ [1, 2, 3, 4]
print(unique_nums)
```
