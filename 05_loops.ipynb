# Loops in Python

Loops allow you to execute a block of code repeatedly. Python provides two primary types of loops:

- **For Loop**: Iterates over items in a sequence (such as a list, string, range, etc.).
- **While Loop**: Repeats as long as a specified condition evaluates to `True`.

## Loop Control Keywords

| Keyword | Use Case                          |
|---------|-----------------------------------|
| `break` | Exit the loop prematurely         |
| `continue` | Skip to the next iteration        |
| `pass`   | Do nothing (acts as a placeholder)|
| `else`   | Execute if the loop completes without a `break` |

#for loop with else block

Python's for loops can optionally include an else block. This else block is executed only if the for loop completes all its iterations without encountering a break statement.

```python
for i in range(10):
    if i == 3:
        break
    print(i)
else:
    print("loop completed")
```
