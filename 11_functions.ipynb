## Function Fundamentals

### Function Definition
- Syntax: `def function_name(parameters):`
- Body contains the code to execute
- Ends with optional `return` statement

### Calling Functions
- Invoke by name followed by parentheses: `function_name(arguments)`
- Arguments must match parameters in number and type (unless defaults are used)

### Default Arguments
- Parameters can have default values: `def func(param=default_value):`
- Defaults must come after non-default parameters
- Allows optional arguments when calling the function

### Variable Arguments
- `*args` for positional arguments (tuple)
- `**kwargs` for keyword arguments (dictionary)

##  Lambda Function 

In Python, a **lambda function** is an anonymous, single-expression function defined using the `lambda` keyword.

Unlike regular functions defined with `def`, lambda functions:
- Do not have a name
- Are limited to a single expression
- Automatically return the result of that expression
---

### Syntax
```python
lambda arguments: expression

Key Points
A lambda function can take no or any number of arguments
It can have only one expression

They can only contain one expression, not multiple statements 

The expression is implicitly returned

Expression can be a number but argument can't be a number
lambda 1: 1 + 1  # ❌ SyntaxError: invalid syntax

Lambda functions are nameless, which means they cannot be directly referenced by a name after their definition unless assigned to a variable.

They can only be reused if assigned to a variable

They do not support complex control flow like full if/else blocks
However, conditional expressions are allowed within the single expression




````python
#basic lambda function
num = 5
x = lambda num: num * 2
print(x(num))
````

````python
 # lambda function in Python can have no arguments.
say_hello = lambda: "Hello!"
print(say_hello())  # Output: Hello!
````

````python
def myfunc(n):
    return lambda a: a * n


#return value i.e lambda function here
mydoubler = myfunc(2)
print(mydoubler)
mytripler = myfunc(3)

#lambda function called
print(mydoubler(11))
print(mytripler(11))
````

````python
# If you don’t call the lambda, it just returns the function object:
say_hi = lambda: "Hi!"
print(say_hi)  # <function <lambda> at 0x...>
print(say_hi())  # Hi!

#if you use print directly inside lambda it'll print the output but willn't return anything
say_hi1 = lambda: print("print value !")

#returns None
print(say_hi1())
````

## Filter Function

- The `filter()` function in Python is a built-in function used to select elements from an iterable based on a condition.
- It keeps only those elements for which the function returns `True`.
- It loops over items in the iterable and filters based on the condition.
- It returns a filter object that must be converted to another iterable like `list`, `tuple`, or `set`.


````python
numbers = [1, 2, 3, 4, 5, 6, 7, 8, 5, 45, 6]
even_numbers = set(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
````

********REDUCE FUNCTION************
To calcuate sum or product of all items in a iterable,instead of large code you have just use reduce function

The reduce() function takes a list and reduces it to a single value by repeatedly applying a function to pairs of elements.


from functools import reduce

reduce(function, iterable)

function: a function that takes 2 arguments.

iterable: list/tuple/etc.

# reduce(f, [a, b, c, d])
# becomes:
f(f(f(a, b), c), d)





````python
from functools import reduce
numbers=[1,2,3,4,5]
sum=reduce(lambda x,y : x+y,numbers)

print(sum)
````

## `*args` and `**kwargs`

- positional arguments: `*args`
- keyword arguments: `**kwargs`

They provide flexibility when you don't know in advance how many arguments a function might receive.

Use them when you want to create functions that can handle a variable number of inputs without modifying the function's signature.

Order must be:

1. normal arguments
2. `*args`
3. `**kwargs`

## Positional Arguments

- **Purpose**: Used to store variable positional arguments
- **Notation**: Denoted by `*args`
- **Return Type**: Stores values in a tuple form

This allows functions to accept an arbitrary number of positional arguments.

## Keyword Arguments

- **Purpose**: Used to store variable keyword arguments
- **Notation**: Denoted by `**kwargs`
- **Return Type**: Stores values in a dictionary form

This allows functions to accept an arbitrary number of keyword arguments.


````python
def function1(*args, **kwargs):
    print(args)
    print(kwargs)


tup = ("Neha", "Sanjna", 56, 344)
dict = {"age": 21, "Stream": "BCA"}

function1(*tup, **dict)
function1("neha",*tup, age=20, color="lavendar")
````

## enumerate 

The `enumerate()` function in Python is used to loop over an iterable (like a list, tuple, or string) while keeping track of the index of each item.

### Syntax
```python
enumerate(iterable, start=0)

````python
fruits = ['apple', 'banana', 'cherry']

for index, fruit in enumerate(fruits):
    print(index, fruit)

for index, fruit in enumerate(fruits,start=101):
    print(index, fruit)
````
