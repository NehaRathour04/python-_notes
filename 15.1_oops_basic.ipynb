## CLASSES AND OBJECTS

`class` 
    - A class is a blueprint or template used to create objects.

`object` 
    - An object is an instance of a class.

A `class` is a blueprint, and an `object` is the actual thing created using that blueprint.
For example, Car is a class, and my BMW car is an object.


````python
class Car :
    def __init__(self,brand,model) : #constructor being used for  instance variables intialization
        self.brand=brand 
        self.model= model
    
    def full_name(self) : #method
        return f"{self.brand} {self.model}"
        
my_car=Car('tata','safari') 
print("my_car brand :",my_car.brand)

my_new_car = Car('toyota','Corolla')
print("my_new_car brand :",my_new_car.brand)

# object methods

print("my_car:",my_car.full_name())
````

---

## **Inheritance**

Inheritance is a concept where one class acquires the properties and methods of another class.

##### 🔹 Simple Explanation
- Parent class (Base class) → gives features  
- Child class (Derived class) → uses those features  

##### 🔹 Real-Life Example
- Parent → Vehicle  
- Child → Car, Bike  

👉 Car inherits features like speed and engine.
```python

class Parent :
    pass

class child(Parent) :# child class inherit Parent class
    pass
```

`super()` method is used to call methods or constructors of the parent class from the child class. 
it is Mostly used to call parent constructor
`super().__init__(baseclass parameters)`

````python
#car class example 

class ElectricCar(Car) : #Car class inherited
    def __init__(self,brand,model,battery_size) :
        super().__init__(brand,model) 
        self.battery_size = battery_size

    def full_name(self) : #method 
        return f"{self.brand} {self.model} {self.battery_size}"
    


my_elec_car = ElectricCar('tesla','model s','85kwh') 
print(my_elec_car.full_name()) #method overriding runtime poly


````

isinstance() is used to check whether an object belongs to a particular class or a tuple of classes.

isinstance(object, class)

````python
result1= isinstance(my_elec_car,ElectricCar)
print(result1," result1")
result2= isinstance(my_elec_car,Car)
print(result2," result2")
````

`Multiple inheritance` allows a class to inherit properties and methods from more than one parent class.

````python
class A:
    def showA(self):
        print("Class A")

class B:
    def showB(self):
        print("Class B")

class C(A, B):   # multiple inheritance
    pass

obj = C()
obj.showA()
obj.showB()
````

`Multilevel inheritance` is when a class inherits from a class, which itself inherits from another class. IT is a chain of inheritance where a class derives from another derived class.

````python
class A:
    def showA(self):
        print("Class A")

class B(A):
    def showB(self):
        print("Class B")

class C(B):   # multilevel inheritance
    pass

obj = C()
obj.showA()
obj.showB()
````
