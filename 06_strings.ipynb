````python
s="hello,my name Is Neha Rathour"
print("case conversion methods")
print(s.upper())
print(s.lower())
print(s.capitalize()) # first letter of the string
print(s.title())     #each word first letter
print(s.swapcase()) #uppercase<-> lowercase
````

````python
print(s)
print("\tsearch check replace methods")
print("find is in s :",s.lower().find('is'))

print("replace:",s.lower().replace('is','in'))
print(f"A is repeated {s.count("a")} times")
print("starts with H :",s.startswith('H') )
print("ends with r :",s.endswith('r') )
````

index() vs slice()
slice() when you want to get a range of characters in a sequence.
here you are writing index numbers
written as s.slice(start,stop,step)
or directly as s[start:stop:step]
s.slice[1,4,2] returns el

Use index() when you want to find position of a value  in a sequence.
here you are writing the sequence
like you want to find index of e in hello so you'll write as s.index('e') returns 1

index() vs find()
index() gives value error if given value not present in string
find() returns -1  if given value not present in string

````python
s1='hello'
print(s1.find('he'))
print(s1.index('he'))

print(s1[slice(1, 4,2)])
print(s1[1:4:2])

````
