## Collection

A **collection** is a group of multiple values stored in a single variable.

### Main Collection Types
- List  
- Tuple  
- Set  
- Dictionary  

----
## Iterables

An **iterable** is any Python object that can be looped over (iterated), typically using a `for` loop.

### Common Examples
- List  
- Tuple  
- String  
- Dictionary  
- Set  
- File  
- Range  

### Note
- `int` is **not iterable**
- `string` **is iterable**

---
## Iterator

In Python, an **iterator** is an object that allows traversal through a collection of data (like lists, tuples, or dictionaries) one element at a time.

It uses the `iter()` and `next()` methods.

---

### Rules of an Iterator
An object is an iterator if it has:

- `__iter__()` → returns the iterator object itself  
- `__next__()` → returns the next item  
  - Raises `StopIteration` when no items are left  

---

### Important Point
- A `for` loop uses iterators internally

```python
list = [2, 7]  #iterable

it = iter(list)
print(type(it))
print(next(it))
print(next(it))
#will give error as  no item left to iterate
# print(next(it))

```

## Generator Function

A **generator function** returns a generator object which can be iterated over (in a `for` loop or using `next()`) until it is exhausted, raising `StopIteration`.

---

### Key Points
- All generators are **iterators**, but not all iterators are generators  
- Generator functions **remember only what is needed** to produce the next value  
- Values are generated **one at a time (lazily)**  
- They are **memory efficient**, making them ideal for large or infinite sequences  

---

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i  # Yields a value and pauses execution
        i += 1

def func():
    result = []
    for i in range(20):
        result.append(i)
   #yielding the whole list
    yield result


n=func()
print(type(n))
print(n)
#returns whole list bcoz yield is called once..and all data stored there
print(next(n))


my_generator = count_up_to(5)
print(type(my_generator))
print(my_generator)
print(next(my_generator))
print(next(my_generator))
```

return: Immediately exits the function and sends a value (or None) to the caller. Once returned, the function's state is gone.


yield: Turns a function into a generator, pausing execution each time it's used, preserving local variables and position. You can resume the function later.
