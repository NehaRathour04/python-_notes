## **Decorator**

A decorator is a function that wraps another function to add extra behavior without modifying the original function's code.

Think of it like a gift wrapper — the gift (function) stays the same, but the wrapper adds something extra around it.

**syntax**

```python

def decorator(func) :
    def wrapper(*args,**kwargs):
        return func(*args,*kwargs)
    return wrapper


@decorator_name
def my_function():
    pass

```

## **Key Points to Remember**

| Concept | What it means |
|----------|--------------|
| `@decorator` | Wraps a function with extra logic |
| `wrapper(*args, **kwargs)` | Makes decorator work with any function signature |
| `return wrapper` | Returns the new wrapped function |
| Stacking `@dec1 @dec2` | You can apply multiple decorators |

*for example*
adding  functionality of time taken by each function using time taken decorator

````python
# write a decorator that measure the
#  time a function takes to execute
````

````python
import time

def time_taken(func) :
    def wrapper(*args,**kwargs):
        start = time.time()
        result=func(*args,**kwargs)
        end=time.time()
        total_time_taken = end - start
        print(f"total time taken by func {func.__name__} {total_time_taken}")
        return result
    return wrapper

@time_taken
def sum(*args):
    sum=0
    for arg in args:
        sum=sum+arg
    time.sleep(4)
    return sum

@time_taken
def example_function(n):
    time.sleep(n)
    
print("example1")
print("sum :",sum(2,34,4,56,4,6,4,6,35,6,4))

print("example2 ")
example_function(3)
````

````python
def debugger(func):
    def wrapper(*args,**kwargs) :
        result=func(*args,**kwargs)
        arg_values=(', '.join(str(arg) for arg in args)) 
        kwargs_values=(", ".join( f"({k} = {v})" for k,v in kwargs.items())) 
        print(f"calling function {func.__name__} with args {arg_values} and kwargs {kwargs_values}")
        return result
    return wrapper

````

````python
@debugger
def total_(*args):
    total_=0
    for arg in args:
        total_=total_+arg
    return total_

total = total_(1,2,3,4)
print(total)
total = total_(3,4)
print(total)
````

## **Dunder Methods (Magic Methods)**

Dunder methods (double underscore methods) are special methods in Python that start and end with double underscores. They are used to define the behavior of objects.

Dunder methods are called internally by Python

some of them __init__ , __add__ ,__str__ , __name__ , __repr__

### 🔹 Examples
```python
1 + 2        # Python calls int.__add__(1, 2)
len("hello") # Python calls str.__len__("hello")
print(obj)   # Python calls obj.__str__()
```

### 🔹 How it Works
When you write:
```python
a + b
```

Python internally calls:
```python
a.__add__(b)
```

````python
class Person:
    def __init__(self, name, age):   # called when object is created
        self.name = name
        self.age = age

    def __str__(self):               # for print() / str()
        return f"Person: {self.name}"

    def __repr__(self):              # for debugging, repr()
        return f"Person('{self.name}', {self.age})"

p = Person("Alice", 30)
print(p)       # Person: Alice       ← uses __str__
repr(p)        # Person('Alice', 30) ← uses __repr__
````
