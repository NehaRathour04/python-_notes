## File Handling

File handling is an essential part of Python programming. It allows you to create, read, update, and delete files. Python provides built-in functions to work with files.

### Open
```python
file = open("example.txt", "mode")
```

Modes:
- 'r' → Read  
- 'w' → Write  
- 'a' → Append  
- 'b' → Binary  
- 'x' → Create  
- 'r+' → Read + Write  

### Read
```python
# Read entire file
content = file.read()

# Read line by line
line = file.readline()

# Read all lines into a list
lines = file.readlines()

```

### Write
```python
file.write("Hello\n")
file.writelines(["Line1\n", "Line2\n"])
```

### Close
```python
file.close()
```

### With Statement
closing file with context mananger
```python
Automatically closes the file after block execution

with open("example.txt", "r") as file:
    content = file.read()
```

````python
#read

with open("DEMOTEXTFILE.txt","r") as file :
    content=file.read()
    print(content)

with open("DEMOTEXTFILE.txt","r") as file1 :
    line=file1.readline()
    print("line :: ",line)

    while(line!=""):
        print(line)
        line =file1.readline()
    print("lines looping end here \n")

#in the form of list
with open("DEMOTEXTFILE.txt","r") as file2 :
    lines=file2.readlines()
    print("lines read through readlines :: ",lines)
````
