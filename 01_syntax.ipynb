---------------Compile time vs run time--------------

compile time - time when code is converted/translates into machine code
during compile time,syntax errors ,undefined variables,type errors are checked
happens before program execution

run time - The time after the program starts running (after successful compilation).
 time when bytecode or machine code actuallly starts executing
during run time,division by zero,array out of bounds,file not found errors are detected

how python checks for compile time errors ??
python read and compile code to bytecode first- but only enough to check for syntax errors,not logical or run time issues. then it executes the code line by line using interpreter

````python
#4. f-string
# formatted string literal 
# allows you to embedd expression inside string literals using curly braces
# allows expressions, not just variables
#To specify a string as an f-string, 
# simply put an f in front of the string literal, and add curly brackets {} as placeholders for variables and other operations.

var="Neha"
print(f"hello :{var}")
print(f"sum of {4+5}")

#6. format method
print("Hello {} ,you can do it".format(var))
print("Hello {1} ,you can do it {0}".format(var,"All the best"))
````

__REPL__
REPL stands for read->evaluate->print-> loop
it is a interative environment where 
    python reads the input
    evaluate it
    print the result
    and loop backs for next command

you can use basic python in terminal directly by just typing python and repl will open
its shown as 
    >>>
it is helpful for learning and testing small code

Jupyter Notebook is an advanced repl where 
    You type Python code in a cell (Read)
    You run the cell (Eval)
    It shows output right below the cell (Print)
    You can run more cells (Loop)

So it follows the REPL pattern, just in a GUI-based, notebook-style format.

___JUPYTER NOTEBOOK__
Jupyter Notebook is an open-source, web-based application for creating and sharing interactive computational documents. It allows users to combine code, text, visualizations, and other multimedia elements in a single document. Think of it as a digital notebook where you can write code, explain it with text, and display the results, all in a single, shareable format.
