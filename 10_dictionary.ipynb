# Dictionary

## Definition
A dictionary stores data in **key-value pairs**.

## Characteristics
- **Mutable**: Can be modified after creation
- **Keys are unique**: No duplicate keys allowed
- **Values can be duplicated**: Multiple keys can have the same value
- **Indexed by key**: Not indexed by position number, but by key
- **Ordered**: From Python 3.7+, dictionaries maintain insertion order
- **Denoted by curly braces**: `{}`

## Syntax

# Dictionary Keys and Set Elements

Dictionary keys and set elements must be **immutable and hashable**.

## Valid Types
- `str` (string)
- `tuple`
- `int`
- `float`
- `None`
- `frozenset`

## Invalid Types
- `set` - mutable and unhashable
- `dict` - mutable and unhashable
- `list` - mutable and unhashable

```python
# ✅ Valid
my_dict = {
    "name": "Alice",
    42: "answer",
    (1, 2): "tuple key"
}

print(my_dict[(1,2)])
print(my_dict[42])

# Dictionary with tuple keys
locations = {
    (0, 0): "Origin",
    (1, 2): "Point A",
    (3, 4): "Point B"
}

# Accessing value using the whole tuple
print(locations[(1, 2)])   # Output: Point A



# ❌ Invalid
# bad_dict = {
#     [1, 2]: "list key"   # TypeError: unhashable type: 'list'
# }
```

# Dictionary Functions

## Common Dictionary Methods

| Method | Description | Syntax |
|--------|-------------|--------|
| `keys()` | Returns all dictionary keys | `dict.keys()` |
| `values()` | Returns all dictionary values | `dict.values()` |
| `items()` | Returns key-value pairs as tuples | `dict.items()` |
| `get()` | Safely retrieves value, returns default if key doesn't exist | `dict.get(key, default)` |
| `pop()` | Removes and returns value for a key | `dict.pop(key, default)` |
| `popitem()` | Removes and returns the last key-value pair | `dict.popitem()` |
| `update()` | Merges another dictionary into current one | `dict.update(other_dict)` |
| `clear()` | Removes all items from dictionary | `dict.clear()` |
| `copy()` | Creates a shallow copy of the dictionary | `dict.copy()` |
| `len()` | Returns number of key-value pairs | `len(dict)` |
| `in` | Checks if key exists in dictionary | `key in dict` |



```python
d={'name':'neha_rathour','age':21,'love':'watching shows'}

print(d)
```

```python
# 1. keys() - Returns a view of all dictionary keys
print("keys():", d.keys())
```

```python
# 2. values() - Returns a view of all dictionary values
print("values():", d.values())
```

```python

# 3. items() - Returns key-value pairs as tuples
print("items():", d.items())
```

```python
# 4. get(key, default) - Safely retrieves value, returns default if key doesn't exist
print("get('a'):", d.get('a'))
print("get('z', 'not found'):", d.get('z', 'not found'))
```

```python
# 6. update(other_dict) - Merges another dictionary into current one
d.update({'d': 40, 'e': 50})
print("After update:", d)
```

```python
# 5. pop(key, default) - Removes and returns value for key
print("pop('b'):", d.pop('b', 'not found'))
print("After pop:", d)
```

```python
# 7. clear() - Removes all items from dictionary
d_copy = d.copy()  # Create a copy first
d_copy.clear()
print("After clear:", d_copy)
```

```python
# 8. copy() - Creates a shallow copy of the dictionary
d_duplicate = d.copy()
print("Copied dict:", d_duplicate)
```

```python
# 9. len() - Returns number of key-value pairs
print("len(d):", len(d))
```

```python
# 10. in - Checks if key exists in dictionary
print("'a' in d:", 'a' in d)
print("'z' in d:", 'z' in d)
```
