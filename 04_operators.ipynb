Identity operators are used to compare the memory locations of two objects. They check whether two variables refer to the same object, not whether their values are equal.
two types :
is 
is not

````python
x=[1,2,3]
y=[1,2,3]
z=x
# x and z points to same memory location
if(x is  z) :
    print("true")
else:
    print("false")

#x and y have same values but they don't point to same memory location
if (x is y):
    print("true ,yes x is y")
else :
    print("false ,no x is y")

if(x==y):
    print("x is equal to y")
````

Bitwise operators
compare bits of two numbers
    1. AND &
    return 1 if both bits are 1
    2. OR |
    return 1 if any of the bit is  1
    3. XOR ^
    return 1 if only one of the bit is 1
    4. not ~
    invert all bits
    0 to 1 ,1 to 0
    uses 2 complement logic here
    ~n equals -n - 1
    5 left shift <<
     The << operator inserts the specified number of 0's from the right and let the same amount of leftmost bits fall off:
     General rule => a << b == a * (2 ** b)
     3 << 2 
     shifts the 3's binary form to  left by adding 2 zeroes at the end
     00000011 << 2 → 00001100
    6. right shift >>
    The right shift operator (>>) shifts the bits of a number to the right by a specified number of positions. For each shift, the rightmost bit is removed, and a 0 is added on the left (for positive numbers).

    General rule => a >> b  ==  a // (2 ** b)

    00000011 >> 2 → 00000000



````python
a=5
b= 3
print("bitwise operations on 5 and 3")
print("and :",a & b)
# 101 for 5
# 011 for 3
# 001 output 1
print("or :", a | b)
# 101 for 5
# 011 for 3
# 111 output 7

print("XOR:",a^b)
# 101 for 5
# 011 for 3
# 110 output 6
print("not :",~a)
 ~n equals -n - 1 
~5 = -5-1=-6
````
