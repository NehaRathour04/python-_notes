a comprehensive example showing most of the oops concept examples

```python
class Car :
    total_car=0 #class attribute
    def __init__(self,brand,model,rating=None) : #constructor being used for  instance variables intialization
        self._brand=brand  #protected var
        self.model= model #object attribute
        self.__rating = rating  #private
        Car.total_car+=1
    
    def full_name(self) : #method
        return f"{self._brand} {self.model}"
     
    def fuel_type(self):
        return "petrol or diesel"
    
    def get_rating(self):
        return self.__rating

    def set_rating(self,rating) :
        self.__rating =rating
    
    @property
    def get_rating_p(self):
        return self.__rating

    @get_rating_p.setter
    def set_rating_p(self,rating) :
        self.__rating =rating

    @classmethod
    def get_total_cars(cls):
        return cls.total_car
    
    def __str__(self) :
        return f"Car {self._brand}"


class ElectricCar(Car) : #Car class inherited
    def __init__(self,brand,model,battery_size) :
        super().__init__(brand,model) 
        self.battery_size = battery_size

    def full_name(self) : #method 
        return f"{self._brand} {self.model} {self.battery_size}"
    
    def fuel_type(self):
        return "electric"
    
    @staticmethod
    def general_description():
        return "cars are means of transport"

   


my_tesla = ElectricCar('tesla','model s','85kwh') 
my_safari= Car("tata","Safari")
print("dunder method __str__ :  ",my_safari)
#polymorphism
print(my_safari.fuel_type()) #instance method
print(my_tesla.fuel_type())

#getter setter
my_tesla.set_rating(5)
print("get rating :",my_tesla.get_rating())

my_tesla.set_rating_p=4
print("get rating using @property :",my_tesla.get_rating_p)


#class variable
print("total car :",my_safari.total_car)

#static method
print("obj access :",my_tesla.general_description())
print("class access to static method :",ElectricCar.general_description())

#class method
print("total cars using class method:", Car.get_total_cars())
```
