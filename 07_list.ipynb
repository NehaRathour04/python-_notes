## List Methods

| Method | Description |
|--------|-------------|
| `append(item)` | Adds item at the end of the list |
| `insert(index, item)` | Adds item at the specified index |
| `pop(index)` | Removes and returns item at the given index |
| `remove(item)` | Removes the first occurrence of the item |
| `clear()` | Removes all items from the list |
| `extend(iterable)` | Adds each element from an iterable individually |
| `sort(reverse=True/False)` | arrange the list in asc or desc order. The data type of list elments should be same all str or all number |
| `count(item)` | count the number of specific items in the list |
| `index(item)` | returns the index of first occurance of the item |

### append() vs extend()

| Method | Behavior | Example |
|--------|----------|---------|
| `append()` | Adds the whole object as a single element | `[1, 2].append([3, 4])` → `[1, 2, [3, 4]]` |
| `extend()` | Adds each element individually | `[1, 2].extend([3, 4])` → `[1, 2, 3, 4]` |

````python
l1 = [1, 2, 3, "neha", 4, 5, "rathour", True]
print("My list : ", l1)
l1.append("appended here")
print("Appended ", l1)
l1.insert(3, "hey")
print("inserted at 3rd index ", l1)
l1.remove(4)
print("element 4 removed :", l1)
l1.pop(4)
print("element at 4th index removed : ", l1)
l1.clear()
print("cleared the list : ", l1)
b = (2, 3, 4, "neha")
l1.append(b)
print(l1)
l1.extend(b)
print(l1)

l1.pop()
print("remove last item : ", l1)

l1.reverse()
print("list reversed ", l1)
````

## sort vs sorted

| `sort()`                   | `sorted()`                                        |
| -------------------------- | ------------------------------------------------- |
| Modifies the original list | Returns a new sorted list                         |
| Works only on lists        | Works on any iterable (list, tuple, string, etc.) |
| No return value (`None`)   | Returns sorted result                             |


````python
nums=[2,4,6,3,5]
nums.sort()  # no return values

print("after sort() :",nums)
````

````python
nums2=[4,6,3,5,7]
sorted(nums2) # return value which has to be saved
````

````python
result= sorted(nums2)
print("Using sorted():", result) # new sorted list 
print("Original list:", nums2) # remains unchanged
````

````python
#copy() vs reference

#any changes to l2 affects l1 and vice versa
# reference to same memory object
l2 = l1

#a separate copy created somewhere in the program
#changes made in l3 have no effect on l1 and vice versa
l3 = l1.copy()
print("l2 : ", l2)
print("l3 :", l3)
if (l2 is l1):
    print("l2 is l1")

if (l3 is l1):
    print("l3 is l1")
````

# split() Method

The `split()` method is used to break a string into a list of substrings based on a specified delimiter.

## Syntax
`string.split(separator, maxsplit)`

**Parameters:**
- `separator` (optional): The delimiter used to split the string. Default is whitespace.
- `maxsplit` (optional): Maximum number of splits. Default is -1 (all occurrences).

**Returns:** A list of substrings.

## Examples
- `'a,b,c'.split(',')` → `['a', 'b', 'c']`
- `'hello world'.split()` → `['hello', 'world']`
- `'one,two,three'.split(',', 1)` → `['one', 'two,three']` (maxsplit limits splits to 1)

## Key Points
- Without arguments, `split()` splits on whitespace and removes empty strings
- Returns a list where each element is a substring
- Useful for parsing CSV data, breaking sentences into words, etc.


````python
# Split string into list of substrings
text = 'one,two,three'
parts = text.split(',', 1)  # ['one', 'two', 'three']
print(parts)
````

# join() Method

The `join()` method is used to concatenate elements of an iterable (like a list) into a single string, using the string on which the method is called as the separator.

## Syntax
`separator.join(iterable)`

## Parameters
- `separator`: The string that will be used to join the elements of the iterable.
- `iterable`: A sequence (e.g., list, tuple) of strings. 

## All elements in the iterable must be strings.

## Returns
A single string with elements joined by the separator.

## Examples
- `' '.join(['Hello', 'World'])` → `'Hello World'`
- `','.join(['a', 'b', 'c'])` → `'a,b,c'`
- `'-'.join(['2024', '01', '15'])` → `'2024-01-15'`

## Key Points
- All elements in the iterable must be strings; use `map()` to convert if needed
- More efficient than concatenating strings with `+` in loops
- The separator can be any string, including empty string `''`

````python
# Join list of strings
words = ['Hello', 'World']

words1 = ('hii', 'hello')
sentence1 = ' '.join(words1)
print(sentence1)
sentence = ' '.join(words)  # 'Hello World'
print(sentence)
print("type of sentence ", type(sentence))

````

## List Comprehension (Python)

List comprehension is a short and simple way to create a new list from an existing iterable (like list, tuple, string).

### Syntax
```python
new_list = [expression for item in iterable if condition]


Explanation
expression → What you want to do with each item
item → Each element from the iterable
iterable → The original list/sequence
condition (optional) → Filter items

````python
nums = [1, 2, 3, 4]

# Square of each number
squares = [x**2 for x in nums]
print(squares)
# Only even numbers
evens = [x for x in nums if x % 2 == 0]
print(evens)
````
