# Python Array Module (Detailed)

This notebook explains the Python array module with definitions, examples, and explanations.

## 📌 What is an Array?
An array is a data structure that stores multiple values of the **same data type** in a single variable.

👉 Unlike lists, arrays are **memory efficient** because they store elements in a compact form.

## 📌 Importing the array module

```python
import array
```

## 📌 Creating an Array

**Syntax:**
`array.array(typecode, iterable)`

- typecode → defines data type
- iterable → initial values

```python
arr = array.array('i', [1, 2, 3, 4])
arr
```

## 📌 Type Codes

| Code | Meaning |
|------|--------|
| i | Integer |
| f | Float |
| d | Double |
| b | Signed char |
| B | Unsigned char |
| u | Unicode character |

## 📌 Accessing Elements
We use indexing like lists.

```python
arr[0],arr[-1],arr[2]
```

## 📌 Adding Elements
- append() → adds one element
- extend() → adds multiple elements

```python
arr.append(5)
arr.extend({6, 7})
arr
```

## 📌 Inserting Elements
Insert element at specific position.

```python
arr.insert(1, 10)
arr
```

## 📌 Removing Elements
- remove() → removes by value
- pop() → removes last element

```python
# arr.remove(10)
arr.pop(3)
arr
```

## 📌 Traversing Array
Loop through elements.

```python
for x in arr:
    print(x)
```

## 📌 Searching
Use index() to find position of element.

```python
arr.index(3)
```

## 📌 Updating Elements
Modify values using index.

```python
arr[0] = 100
arr
```

## 📌 Array Methods
- reverse() → reverse array
- count() → count occurrences

```python
arr.reverse()
arr.count(2), arr
```

## 📌 Memory Efficiency
Arrays use less memory than lists because they store elements in compact form.

```python
arr.buffer_info()
```

## 📌 Convert Array to List
Sometimes we convert array to list for flexibility.

```python
arr.tolist()
```

## 📌 Advantages of Array
✅ Memory efficient
✅ Faster for numeric data
✅ Simple structure

## 📌 Limitations of Array
❌ Only one data type allowed
❌ Less flexible than lists
❌ Not suitable for complex operations

## 📊 List vs Array vs NumPy (Quick Revision)

| Feature | List | Array | NumPy |
|--------|------|------|-------|
| Data type | Mixed | Same | Same |
| Speed | Slow | Medium | Fast |
| Memory | High | Less | Very less |
| Dimensions | 1D | 1D | Multi-D |
| Math ops | No | Limited | Yes |

### 🔹 Use
- **List** → general use  
- **Array** → memory-efficient numbers  
- **NumPy** → fast calculations / data science
