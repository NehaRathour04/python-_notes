# Abstraction in Python

Abstraction means **hiding complex implementation details** and only exposing what's necessary to the user.

> Think of a TV remote — you press a button and it works. You don't need to know the electronics inside. That's abstraction.

## How to Achieve Abstraction in Python?

Using the **`abc` module** (Abstract Base Classes)

- Inherit from `ABC`
- Mark methods with `@abstractmethod` that **must** be implemented by subclasses

```python
from abc import ABC, abstractmethod
```

## Simple Example — Animal

```python
from abc import ABC, abstractmethod

class Animal(ABC):          # Abstract class
    
    @abstractmethod
    def sound(self):        # Abstract method - no body
        pass

    @abstractmethod
    def move(self):
        pass

    def breathe(self):      # Normal method - has body
        print("Inhale... Exhale...")  # shared for all animals
```

```python
class Dog(Animal):
    def sound(self):
        print("Woof!")
    
    def move(self):
        print("Runs on 4 legs")

class Bird(Animal):
    def sound(self):
        print("Tweet!")
    
    def move(self):
        print("Flies with wings")

d = Dog()
d.sound()    # Woof!
d.breathe()  # Inhale... Exhale...
```

## What if you DON'T implement abstract methods?

Python **forces** you to implement all abstract methods. If you forget even one, it throws a `TypeError`.

```python
class Cat(Animal):
    def sound(self):
        print("Meow!")
    # forgot to implement move()

try:
    c = Cat()  # TypeError: Can't instantiate abstract class
except TypeError as e:
    print(f"Error: {e}")
```

## You Can't Instantiate an Abstract Class

It's just a **blueprint** — not meant to be used directly.

```python
try:
    a = Animal()  # TypeError
except TypeError as e:
    print(f"Error: {e}")
```

## Real World Example — Payment System

```python
from abc import ABC, abstractmethod

class Payment(ABC):

    @abstractmethod
    def pay(self, amount):     # every payment method must have this
        pass

    @abstractmethod
    def verify(self):
        pass

class CreditCard(Payment):
    def pay(self, amount):
        print(f"Paid ₹{amount} via Credit Card")
    
    def verify(self):
        print("Verifying card details...")

class UPI(Payment):
    def pay(self, amount):
        print(f"Paid ₹{amount} via UPI")
    
    def verify(self):
        print("Verifying UPI PIN...")

# Usage - user doesn't care HOW payment works internally
def process_payment(payment: Payment, amount):
    payment.verify()
    payment.pay(amount)

process_payment(CreditCard(), 500)
process_payment(UPI(), 200)
```

## Abstract Property

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @property
    @abstractmethod
    def area(self):       # subclasses must define area
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    @property
    def area(self):
        return 3.14 * self.radius ** 2

c = Circle(5)
print(c.area)   # 78.5
```

## Abstract Methods — No Body!

Abstract methods have **no body** — just `pass`. The abstract class is a blueprint:

> "Every subclass MUST have this method — but I don't care how you implement it, that's your job."

Technically, abstract methods *can* have a body, but it's rare. Subclasses still must override them, and can optionally call `super()` to use the parent body.

```python
# Abstract method with a body (rare but valid)
class Animal(ABC):
    @abstractmethod
    def sound(self):
        print("Some generic animal sound")  # has a body

class Dog(Animal):
    def sound(self):
        super().sound()    # optionally call parent body
        print("Woof!")

Dog().sound()
```

## Abstraction vs Encapsulation

| | Abstraction | Encapsulation |
|---|---|---|
| **What** | Hides complexity | Hides data |
| **How** | Abstract classes / methods | Private variables (`_`, `__`) |
| **Focus** | What an object *does* | How data is *protected* |
| **Example** | `pay()` hides payment logic | `__balance` hides bank balance |

## Key Takeaways

- Abstraction = **show only what's needed, hide the rest**
- Use `ABC` and `@abstractmethod` to enforce it
- Abstract classes act as **contracts/blueprints** for subclasses
- You **cannot** instantiate an abstract class directly
- Abstract methods have **no body** — just `pass`
- Forces **consistency** across all subclasses
