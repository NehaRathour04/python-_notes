__________PYTHON DATA TYPES_________

```python
Python is a dynamically typed language 
where variable types are decided at runtime
You don’t need to declare data type of a variable

Python is strongly typed language because implicit type conversion is not possible and it doesn't auto convert and throws a error instead
print("Age: " + 25)  # TypeError ❌
you have to first explicitly convert it : 
print("Age: " + str(25))  # ✅
```

```python

Data types define kind of value a variable holds

**detailed data types**

Numeric Types:	int, float, complex
Text Type:	str
Sequence Types:	list, tuple, range
Mapping Type:	dict
Set Types:	set, frozenset
Boolean Type:	bool
Binary Types:	bytes, bytearray, memoryview # not covered yet
None Type:	NoneType
```

```python
1.Numeric type
    *int*       23
    for whole numbers(positive,negative,zero)
    *float*     45.78
     floating or decimal numbers
    *complex*   2+3i

 complex numbers have two parts   real and imaginery
real+imaginery => 2 + 3i
useful in maths ,electrical engineering
you can also declare it as a=complex(2,3) output will be 2+3i

2.text type
    *str*  "hello" 'hello' '''hello'''
    string is the sequence of characters
    immutable

3.boolean type
    *bool*
    have two types True or False(T and F must be in capitaal letters )
    used in conditions,logics and comparisons
    if(a>b) : here a boolean value True or False is returned depending on which block of code is executed

4.sequence type
    *list*
    ordered,mutable,allow duplicates, in []
    [2,3,4,5]
    *tuple*
    ordered,immutable,allow duplicates, in ()
    (2,3,4,5)
    *range*
    immutable sequence of numbers
    used in loops
5. set types
    *set*
    collection of elements
    unordered mutable unindexed and allow no duplicate values
    *frozenset*
    same as set but immutable
6 mapping type
   *dict*
    stores data in key value pairs
    unordered mutable
7 NoneType
    *None*
    represents no value
    like x= null



To check the data type of variable , type() function is used
and for type conversion its simply done as name_type(variable)
here name_type means the type you want your variable to change to

```
