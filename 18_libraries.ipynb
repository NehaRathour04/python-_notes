# 3. SHUTIL module
 for high level files and directory operations
 such as copying moving deleting files or folders
| Function       | Purpose                 |
| -------------- | ----------------------- |
| `copy()`       | Copy file               |
| `copy2()`      | Copy file with metadata |
| `copytree()`   | Copy entire directory   |
| `move()`       | Move file or directory  |
| `rmtree()`     | Delete directory        |
| `disk_usage()` | Show disk usage         |


```python
# shutil main functions

# copy a file
shutil.copy(src, dst)           # metadata not preserved
shutil.copy2(src, dst)          # metadata copied

# copy just the file contents
shutil.copyfile(src, dst)

# copy an entire directory tree
shutil.copytree(src_dir, dst_dir)

# move/rename a file or directory
shutil.move(src, dst)

# remove a directory tree
shutil.rmtree(path)

# create and unpack archives
shutil.make_archive(base_name, format, root_dir)
shutil.unpack_archive(filename, extract_dir=None)

# other utilities
shutil.disk_usage(path)         # usage statistics
shutil.chown(path, user=None, group=None)
```

# 4. RANDOM MODULE

## Important Functions of Python Random Module

| Function | Description | Example |
|---|---|---|
| `random()` | Returns a random float between 0.0 and 1.0 | `random.random()` |
| `randint(a, b)` | Returns a random integer between a and b (inclusive) | `random.randint(1,10)` |
| `randrange(start, stop, step)` | Returns a random number from a range | `random.randrange(1,10,2)` |
| `choice(seq)` | Returns one random element from a sequence | `random.choice(list)` |
| `choices(seq, k=n)` | Returns multiple random elements (with repetition) | `random.choices(list, k=3)` |
| `sample(seq, k)` | Returns unique random elements (no repetition) | `random.sample(list,3)` |
| `shuffle(list)` | Shuffles the elements of a list randomly | `random.shuffle(list)` |
| `uniform(a, b)` | Returns a random float between a and b | `random.uniform(1,10)` |
| `seed(n)` | Initializes the random generator seed() helps to repeat the same random numbers.| `random.seed(5)` |
| `getrandbits(k)` | Returns an integer with k random bits | `random.getrandbits(8)` |

```python
import random
```

```python
random.sample([2,3,4,5,6,7],3)
```

```python
random.choices([2,3,4,5,6,7],k=3)
```
