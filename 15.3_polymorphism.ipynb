## **Polymorphism**

The word *polymorphism* means "many forms", and in programming it refers to methods, functions, or operators with the same name that can be executed on different objects or classes.
-  For example, the + operator is used to add numbers and also to concatenate strings.

---

### 🔹 Operator Overloading
- Same operator behaves differently with different data types  

---

### 🔹 Method Overriding (Runtime Polymorphism)
- Runtime polymorphism means that the behavior of a method is decided while the program is running, based on the object calling it  
- Example: A child class overrides a parent class method  

---

### 🔹 Method Overloading (Compile-Time Polymorphism)
- Same method name but different arguments  
- Decides which method to run at compile time  
- Not directly supported in Python, but can be implemented using `*args` and `**kwargs`


super

```python
class Parent:
    def __init__(self, name):
        self.name = name
        print(f"Parent __init__ called for {self.name}")

    def greet(self):
        print(f"Hello from Parent, {self.name}!")

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)  # Call parent's __init__
        self.age = age
        print(f"Child __init__ called for {self.name}, age {self.age}")

    def greet(self):
        super().greet()  # Call parent's greet method
        print(f"Hello from Child, {self.name} of age {self.age}!")

# Create an instance of the Child class
child_obj = Child("Alice", 10)

# Call the greet method  
child_obj.greet() # method overriding
```

```python
## another example runtime polymorphism

class Car :
    brand=None  # attribute
    model=None
    def __init__(self,brand,model) : #constructor being used for  instance variables intialization
        self.brand=brand 
        self.model= model
    
    def full_name(self) : #method
        return f"{self.brand} {self.model}"
     
    def fuel_type(self):
        return "petrol or diesel"


class ElectricCar(Car) : #Car class inherited
    def __init__(self,brand,model,battery_size) :
        super().__init__(brand,model) 
        self.battery_size = battery_size

    def full_name(self) : #method 
        return f"{self.brand} {self.model} {self.battery_size}"
    
    def fuel_type(self):
        return "electric"



my_tesla = ElectricCar('tesla','model s','85kwh') 
my_safari= Car("tata","Safari")

print(my_safari.fuel_type())
print(my_tesla.fuel_type())

```
