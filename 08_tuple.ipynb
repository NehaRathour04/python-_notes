### Operations on Tuples

- **Slicing**: Extract a portion of the tuple using indices (e.g., `a[1:4]`).
- **Indexing**: Access elements by their position (e.g., `a[3]`).
- **Concatenation**: Combine tuples using `+` (e.g., `t1 + t2`).
- **Repetition**: Repeat elements using `*` (e.g., `t1 * 3`).
- **Membership Test**: Check if an item is in the tuple using `in` (e.g., `3 in t1`).
- **Iteration**: Loop through elements using a `for` loop.


````python
a=(2,3,4,5,6,7)
print(a[1:4])
print(a[3])

t1=(2,3,4)
t2=(5,6,7)
n1,n2,n3=t1
print(f"t1 unpacked \n n1 :{n1},n2:{n2},n3:{n3}")
print(t1+t2)
print("repition : ",t1*3)
print("membership operation : ",3 in t1)
print("loop ")
for i in a :
    print(i)
````

### Tuple Methods

- **`index(item)`**: Returns the index position of the first occurrence of the specified item in the tuple.
- **`index(value, start=0, end=len(tuple))`**: Returns the first index where the value appears, starting from the optional `start` index (default 0) up to the optional `end` index (default length of the tuple).
- **`count(item)`**: Returns the count of occurrences of the specified item in the tuple.
- **`tuple()` constructor**: Creates a new tuple from an iterable (e.g., list, string) or an empty tuple if no arguments are provided.


````python
t1=(3,4,2,5,5,6,2,1)
print("count of 2 in t1",t1.count(2))
print("index of 3 in tuple t1",t1.index(3))
````
