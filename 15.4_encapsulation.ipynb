## **Encapsulation (Data Hiding)**

Encapsulation is a principle of OOP that binds data and methods together into a single unit (class).  
It allows control over the access and visibility of data (attributes) and methods, helping to protect and organize code.

---

### 🔹 Main Purpose
- To hide internal functionality  
- To restrict direct access to data  

---

### 🔹 Access Modifiers

#### 🔸 Private
- Cannot be accessed outside the class  
- Accessible only inside the class  
- Represented using double underscore: `__var`  

#### 🔸 Protected
- Can be accessed within the class and by child (derived) classes  
- Represented using single underscore: `_var`  

#### 🔸 Public
- Can be accessed from anywhere in the program  

**important point** 

Private attributes cannot be accessed directly using objects, 

while protected attributes can be accessed but are meant for internal use by convention in Python.

```python
#private
class Person:
    def __init__(self, name, age):
        self.__name = name  #cant be accessed outside class directly
        self.__age = age

    def display_info(self):
        print(f"name is {self.__name} and age is {self.__age}")


P1 = Person("Neha", 21)
P1.display_info()

#will throw error
# print(P1.__name)
print(dir(P1))
```

```python
#protected
class Person:
    def __init__(self, name, age):
        self._name = name  #cant be accessed outside class directly
        self._age = age

    def display_info(self):
        print(f"name is {self._name} and age is {self._age}")


#using in derived class
class Person1(Person):
    def __init__(self, name, age):
        super().__init__(name, age)

    def display(self):
        print(f"{self._name}'s age is {self._age}")


P =Person("sanjna",44)
print(P._name)

P1 = Person1("Neha", 21)
#you can access it here directly , but not a good practice
print(P1._name)

P1.display_info()
P1.display()







```

```python
#public
class Person:
    def __init__(self, name):
        self.name = name


P1 = Person("Neha")
print(P1.name)
```

Getters and setters are methods used to access and update private variables in a controlled way.

Explanation
Getter → gets (reads) value
Setter → sets (updates) value

👉 Used when variables are private (__var)

```python
class MyClass:
    def __init__(self, value):
        self.__value = value


c1 = MyClass(23)
#error you can't access private attribute directly
print(c1.__value)


```

*advanced concept of getter setter using @property method*


@property is a decorator that allows you to access a method like an attribute.

```python
class Myclass :
    def __init__(self,value):
        self.__value= value

    @property
    def getvalue(self):
        return self.__value
    
    @getvalue.setter
    def setvalue(self,value):
        self.__value = value

c1=Myclass(21)
print("c1.getvalue : ",c1.getvalue)

c1.setvalue = 45
print("c1.getvalue : ",c1.getvalue)
```

```python
class MyClass:
    def __init__(self, value):
        self.__value = value #private

    @property
    def valuefunc(self):
        return self.__value

c1 = MyClass(23)
print(c1.valuefunc)
```

```python
# setting value using setter

class MyClass:
    def __init__(self, value):
        self.__value = value

    @property
    def valuefunc(self):
        return self.__value
    
    @valuefunc.setter
    def valuefunc(self,new_value):
        self.__value= new_value

c1 = MyClass(23)
print(c1.valuefunc)
c1.valuefunc=30
print(c1.valuefunc)
```
