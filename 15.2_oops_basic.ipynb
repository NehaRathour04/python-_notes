## **Important Point**

- Object attributes (instance variables) are given preference over class attributes.

## **Object Attributes (Instance Variables)**

- Belong to an instance (object) of a class  
- Defined using `self.attribute` inside methods like `__init__`  
- Unique to each object  

```python
class Dog:
    def __init__(self, name):
        self.name = name  # Object attribute

d1 = Dog("Rex")
d2 = Dog("Buddy")

print(d1.name)  # Rex
print(d2.name)  # Buddy
```

## **Class Attributes (class variable)**

- Shared by all instances of the class  
- Defined directly inside the class, outside methods  
- Can be accessed using the class name or `self.__class__.attribute`

```python
class Dog:
    name = "Canine"  # Class attribute

    def __init__(self, name):
        self.name = name  # Object attribute

d1 = Dog("Rex")
d2 = Dog("Buddy")

print(Dog.name)
#object attribute given more preference
print(d1.name)
print(d2.name)

```

## **Static Method**

- A static method is a method that belongs to a class but does not operate on a specific instance or the class itself  
- It is defined using the `@staticmethod` decorator  
- It does not take `self` or `cls` as arguments  
- it is used to define utility functions that belong to a class but do not depend on instance or class data.

```python
class TemperatureConverter:
    @staticmethod
    def celsius_to_fahrenheit(c):
        return (c * 9/5) + 32

t1=TemperatureConverter()
print("object access :",t1.celsius_to_fahrenheit(32) )
print("direct class access : ",TemperatureConverter.celsius_to_fahrenheit(32))
```

## **Instance Method**

- An instance method is a function defined within a class that operates on a specific instance (object) of that class  
- it can't be called using class itself
### 🔹 `self` Parameter
- The first parameter of an instance method is conventionally named `self`  
- It refers to the current instance (object) of the class  
- It allows access to and modification of the instance’s attributes  


> Note: `self` is just a convention; any valid name can be used, but `self` is strongly recommended.

```python
class example :
    #instance method
    def __init__(self,name):
        self.name=name

    def display(self):
        return self.name

e1=example("Neha")
print("called using instance",e1.display())
#throw error
# print("called using class",example.display())
```

## **Class Method**

- A class method is a method that is bound to the class rather than an instance of the class  
- It operates on class-level data (class attributes), not on individual object data  

### 🔹 Key Points
- Defined using the `@classmethod` decorator  
- Takes `cls` as the first parameter, which refers to the class  
- Can be called using the class name or an instance  

### 🔹 Example Usage
- `ClassName.method_name()`  
- `object_name.method_name()`

```python
class classMethdEx :
    @classmethod
    def  class_method_ex(cls,value):
        cls.class_var=value
        # print(cls.class_var)

classMethdEx.class_method_ex(32)
print(classMethdEx.class_var)
```
